```python
n_samples = 10
X_p = np.linspace(0, 1, n_samples)
Y_p = np.linspace(0, 1, n_samples)
X_p, Y_p = np.meshgrid(X_p, Y_p)
X_p = X_p.flatten()
Y_p = Y_p.flatten()
samples = np.array([X_p, Y_p]).T
```

## Example B

$$\frac{\partial^2\Psi}{\partial y \partial x}(x, y)=-4x\exp(-y)$$

The analytic solution is: $\Psi(x,y)=(x-2)(x+2)(2\exp(-y)+3)$.

With boundary conditions $\Psi(0,y)=-4(2\exp(-y)+3)$, $\Psi(1, y)=-3(2\exp(-y)+3)$, $\Psi(x, 0)=5(x+2)(x-2)$, and $\Psi(x, 1)=(x+2)(x-2)(2\exp(-1)+3)$.

The trial solution for this case is $\Psi(x, y)=A(x, y) + B(x, y)N(x, y)$ with $B(x)=x(1-x)y(1-y)$ and $A(x, y)= -4(1-x)(2\exp(-y)+3) -3x(2\exp(-y)+3) +(1-y)(5(x+2)(x-2)+20(1-x)+15x) + y((x+2)(x-2)(2\exp(-1)+3)+4(1-x)(2\exp(-1)+3)+3x(2\exp(-1)+3)).$

```python
def exampleB_boundary_value(point):
  x = point[0]
  y = point[1]
  return (-4*(1-x)*(2*np.exp(-y)+3) -3*x*(2*np.exp(-y)+3)
          +(1 - y)*(5*(x+2)*(x-2) + 20*(1-x) +15*x)
          +y*((2*np.exp(-1)+3)*(x+2)*(x-2) + 4*(2*np.exp(-1)+3)*(1-x) +3*(2*np.exp(-1)+3)*x))
```

```python
def exampleB_boundary_vanishing(point):
  x = point[0]
  y = point[1]
  return x * (1 - x) * y * (1 - y)
```

### Defining the loss function for a single point and a whole set

The loss function is based on the formula:
$$Loss(N)=\sum_{i,j} \left(\frac{\partial^2\Psi}{\partial y \partial x}(x_i, y_j)+4x_i\exp(-y_j)\right)^2$$
And the Laplace operator term acting on the trial solution becomes:
$$\frac{\partial^2\Psi}{\partial y \partial x}=\frac{\partial^2 A}{\partial y \partial x} + \frac{\partial^2}{\partial y \partial x} \left(x(1-x)y(1-y)N(x, y)\right)$$
$$=-2\exp(-y) +(2x-1)(2\exp(-1)-2)+(1-2y)(1-2x)N(x, y) +4x\exp(-y)$$
$$+ (1-2x)y(1-y)\frac{\partial N}{\partial y}(x, y) +
x(1-x)(1-2y)\frac{\partial N}{\partial x}(x, y) + y(1-y)x(1-x)\frac{\partial^2 N}{\partial y\partial x}(x, y)$$

```python
def exampleB_loss_function_single_point(self, point, non_squared=False, *kwargs):
  x = point[0][0] if isinstance(point[0], np.ndarray) else point[0]
  y = point[1][0] if isinstance(point[1], np.ndarray) else point[0]
  
  N = self.forward_pass(point, 0)
  dN = self.forward_pass(point, 1)
  dydxN = self.forward_pass_arbitrary_derivative(point, np.array([1, 1]))
  
  loss = ( (1-2*y)*(1-2*x)*N[0] + (1-2*x)*y*(1-y)*dN[0, 1] +(1-2*y)*x*(1-x)*dN[0, 0] + y*(1-y)*x*(1-x)*dydxN[0]
           -2*np.exp(-y) + (2*x-1)*(2*np.exp(-1)-2) +4*x*np.exp(-y)
    )
  if non_squared:
    return loss
  else:
    loss = loss ** 2
    return loss
```

```python
def exampleB_loss_function(self, samples, *kwargs):
  loss = 0
  for i in range(samples.shape[0]):
    loss += self.loss_function_single_point(self, samples[i])
  return loss/samples.shape[0]
```

### Defining the update rules

The following functions represent $\frac{\partial Loss}{\partial \vec{b}}$, $\frac{\partial Loss}{\partial H}$, and $\frac{\partial Loss}{\partial V}$

```python
def exampleB_bias_change(self, point, label, *kwargs):
  x = point[0][0] if isinstance(point[0], np.ndarray) else point[0]
  y = point[1][0] if isinstance(point[1], np.ndarray) else point[0]
  db = np.zeros((self.hidden_dim, 1)).astype(dtype="float64")
  loss_sqrt = self.loss_function_single_point(self, point, non_squared=True)
  
  db_N = self.network_derivative_bias(point, 0)
  db_DN = self.network_derivative_bias(point, 1)
  db_dydxN = self.arbitrary_network_derivative_bias(point, np.array([1, 1]))
  
  for m in range(self.hidden_dim):
    db[m] += 2 * loss_sqrt * ( (1-2*y)*(1-2*x)*db_N[0, 0, m] + (1-2*x)*y*(1-y)*db_DN[0, 1, m]
                              +(1-2*y)*x*(1-x)*db_DN[0, 0, m] + y*(1-y)*x*(1-x)*db_dydxN[0, m]
      )
  return db
```

```python
def exampleB_hidden_weights_change(self, point, *kwargs):
  x = point[0][0] if isinstance(point[0], np.ndarray) else point[0]
  y = point[1][0] if isinstance(point[1], np.ndarray) else point[0]
  dH = np.zeros((self.hidden_dim, self.input_dim)).astype(dtype="float64")
  loss_sqrt = self.loss_function_single_point(self, point, non_squared=True)
  
  dH_N = self.network_derivative_hidden_weights(point, 0)
  dH_DN = self.network_derivative_hidden_weights(point, 1)
  dH_dydxN = self.arbitrary_network_derivative_hidden_weights(point, np.array([1, 1]))

  
  for m in range(self.hidden_dim):
    for p in range(self.input_dim):
      dH[m, p] += 2 * loss_sqrt * ((1-2*y)*(1-2*x)*dH_N[0, 0, m, p] + (1-2*x)*y*(1-y)*dH_DN[0, 1, m, p]
                              +(1-2*y)*x*(1-x)*dH_DN[0, 0, m, p] + y*(1-y)*x*(1-x)*dH_dydxN[0, m, p]
      )
  return dH
```

```python
def exampleB_visible_weights_change(self, point, *kwargs):
  x = point[0][0] if isinstance(point[0], np.ndarray) else point[0]
  y = point[1][0] if isinstance(point[1], np.ndarray) else point[0]
  dV = np.zeros((self.visible_dim, self.hidden_dim)).astype(dtype="float64")
  loss_sqrt = self.loss_function_single_point(self, point, non_squared=True)
  
  dV_N = self.network_derivative_visible_weights(point, 0)
  dV_DN = self.network_derivative_visible_weights(point, 1)
  dV_dydxN = self.arbitrary_network_derivative_visible_weights(point, np.array([1, 1]))
  
  for m in range(self.visible_dim):
    for p in range(self.hidden_dim):
      dV[m, p] += 2 * loss_sqrt * ((1-2*y)*(1-2*x)*dV_N[0, 0, m, p] + (1-2*x)*y*(1-y)*dV_DN[0, 1, m, p]
                              +(1-2*y)*x*(1-x)*dV_DN[0, 0, m, p] + y*(1-y)*x*(1-x)*dV_dydxN[0, m, p]
      )
  return dV
```

### Defining the trial solution with an apropiate network

```python
exampleB_trial_solution = nnde.TrialSolution(loss_function=exampleB_loss_function,
                                        loss_function_single_point=exampleB_loss_function_single_point,
                                        bias_change=exampleB_bias_change,
                                        hidden_weights_change=exampleB_hidden_weights_change,
                                        visible_weights_change=exampleB_visible_weights_change,
                                        boundary_condition_value_function=exampleB_boundary_value,
                                        boundary_vanishing_function=exampleB_boundary_vanishing,
                                        input_dim=2, hidden_dim=10, output_dim=1, learning_rate=0.0005, momentum=0.2)
```

### Training for 5000 epochs

```python
# %%pixie_debugger
exampleB_trial_solution.train(samples, 5000)
```

### Plotting the results 

The numerical solution (training set - red, valdiaiton set - green) along with the analytical solution (blue).

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
new_shape = int(np.sqrt(samples.shape[0]))
ZeBsol = np.array([exampleB_trial_solution.predict(samples[i]) for i in range(samples.shape[0])]).reshape((samples.shape[0],))
surf = ax.plot_surface(X=samples[:,0].reshape((new_shape, new_shape)), Y=samples[:,1].reshape((new_shape, new_shape)),
                Z=ZeBsol.reshape((new_shape, new_shape)), label='Numerical - Training')
surf._facecolors2d=surf._facecolors3d
surf._edgecolors2d=surf._edgecolors3d# ax.scatter3D(xs=samples[:,0], ys=samples[:,1], zs=ZeBsol, label='Numerical - Training')
plt.legend(loc=2)
plt.show()
```
