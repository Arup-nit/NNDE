```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
import nnde
```

# Test for RMSE

Let's test if the written shallow network class works for a simple case of fitting it to the cosine function with the RMSE loss function.

```python
X = np.arange(0, 2 * np.pi, 0.025)
X = X.reshape((X.shape[0], 1, 1))
Y = np.sin(X)
```

## Defining the netowork for the task

```python
network = nnde.ShallowNetwork(hidden_dim=100, unsupervised=False,
                              momentum=0, learning_rate=0.01, predefined_loss='rmse')
```

Training for $10^3$ epochs.

```python
network.train(samples=X, labels=Y, epochs=5000)
```

Let's plot the results.

```python
Y2 = np.array([network.forward_pass(X[i], 0) for i in range(X.shape[0])]).reshape((X.shape[0],))
Xp = X.reshape((X.shape[0],))
Yp = Y.reshape((Y.shape[0],))
plt.plot(Xp, Yp, label='Ground truth', c='g')
plt.plot(Xp, Y2, label='Predictions', c='b')
plt.legend()
plt.show() 
```

```python
plt.clf()
plt.plot(Xp, Yp, label='Ground truth', c='xkcd:sky blue', lw=10)
plt.plot(Xp, Y2, label='Predictions', c='xkcd:goldenrod', lw=10)
plt.xlabel(r'$x$', fontsize='50')
plt.ylabel(r'$y$', fontsize='50')
# plt.axis('equal')
plt.legend(fontsize='40')
plt.title('Fitting cosine function', fontsize='60')
plt.gcf().set_size_inches(30, 22.5)
plt.tick_params(axis='both', which='major', labelsize=35)
plt.savefig('plots/rmse_cos.jpg')
plt.show()
```
