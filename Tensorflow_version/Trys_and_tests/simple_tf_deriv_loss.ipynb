```python
from __future__ import absolute_import, division, print_function, unicode_literals
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
class ShallowNetwork(tf.keras.models.Model):
  def __init__(self, n_h, activation='sigmoid'):
    super(ShallowNetwork, self).__init__()
    self.activation = activation
    self.n_h = n_h
    self.hidden_layer = tf.keras.layers.Dense(units=self.n_h, activation=self.activation)
    self.output_layer = tf.keras.layers.Dense(units=1, activation='linear')
  
  def call(self, X):
    X = tf.convert_to_tensor(X)
#     print(X[])
    response = self.hidden_layer(X)
    return self.output_layer(response) 
```

```python
net = ShallowNetwork(10)
```

```python
X = np.arange(-1,1,0.01).reshape(-1,1)
Y = 2*X
```

```python
def diff_loss(model, inputs, target_derivatives):
    target_derivatives = tf.convert_to_tensor(target_derivatives)
    with tf.GradientTape() as tape:
      inputs = tf.convert_to_tensor(inputs)
      tape.watch(inputs)
      response = model(inputs)
    grads = tape.gradient(response, inputs)
#     loss = tf.keras.metrics.mean_squared_error(target_derivatives, grads)
    loss = tf.square(target_derivatives - grads)
    cond = tf.square(model(tf.convert_to_tensor(np.array([[0]])))-tf.constant(0.,dtype='float64'))
    loss += cond
    return loss
```

```python
optimizer = tf.keras.optimizers.Adam(learning_rate=0.4)
train_loss = tf.keras.metrics.Mean('train')
```

```python
@tf.function
def train_step(x, y):
  with tf.GradientTape() as tape:
    preds = net.call(x)
    loss= diff_loss(net, x, y)
  gradients = tape.gradient(loss, net.trainable_variables)
  optimizer.apply_gradients(zip(gradients, net.trainable_variables))
  
  train_loss(loss)
```

```python
EPOCHS = 100
for epoch in range(EPOCHS):
  train_step(X, Y)
  print(train_loss.result().numpy())
  if (epoch+1) % 10 == 0:
    pred = net.call(tf.convert_to_tensor(X,dtype='float32')).numpy()
    plt.plot(X, pred)
    plt.plot(X, X**2)
    plt.show()
```
