```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
import nnde
```

# Test for RMSE

Let's test if the written shallow network class works for a simple case of fitting it to the cosine function with the RMSE loss function.

```python
X = np.arange(0, 2 * np.pi, 0.025)
X = X.reshape((X.shape[0], 1, 1))
Y = np.sin(X)
```

## Defining the netowork for the task

```python
network = nnde.ShallowNetwork(hidden_dim=75, unsupervised=False,
                              momentum=0, learning_rate=0.1, predefined_loss='rmse')
```

Training for $10^3$ epochs.

```python
network.train(samples=X, labels=Y, epochs=1000)
```

Let's plot the results.

```python
Y2 = np.array([network.forward_pass(X[i], 0) for i in range(X.shape[0])]).reshape((X.shape[0],))
Xp = X.reshape((X.shape[0],))
Yp = Y.reshape((Y.shape[0],))
plt.plot(Xp, Yp, label='Ground truth', c='g')
plt.plot(Xp, Y2, label='Predictions', c='b')
plt.legend()
plt.show() 
```
